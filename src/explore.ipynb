```python
data = teams.MATCHUP

def determinar_tipo(data):
    if '@' in data:
        return 'A'  # Si hay '@', entonces es local
    elif 'vs' in data:
        return 'H'  # Si hay 'vs', entonces es visitante
    else:
        return None  # Si no coincide ninguno de los patrones, devuelve None


teams['HOME_AWAY'] = teams['MATCHUP'].apply(determinar_tipo)

teams
```

```python
import pandas as pd

# Crear DataFrames de ejemplo
data1 = {'columna1': [1, 2, 3, 4],
         'columna2': ['A', 'B', 'C', 'D']}
df1 = pd.DataFrame(data1)

data2 = {'columna1': [2, 4, 6, 8],
         'columna2': ['B', 'D', 'F', 'H'],
         'valor_df2': [20, 40, 60, 80]}
df2 = pd.DataFrame(data2)
# Combinar los DataFrames basándote en las columnas 'columna1' y 'columna2'
merged_df = pd.merge(df1, df2, on=['columna1', 'columna2'], how='left')

# Agregar los valores de 'valor_df2' a 'df1' si hay un match
df1['valor_df2'] = merged_df['valor_df2']

# Mostrar el DataFrame resultante
print(df1)
```
